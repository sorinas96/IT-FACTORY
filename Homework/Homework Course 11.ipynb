```python
# 1. a) Creati un context manager care genereaza in mod aleatoriu o lungime si o latime pentru un dreptunghi. Calculati si printati aria dreptunghiului in interiorul context managerului.
# b) Modificati codul astfel in cat se se permita rularea si cu valori predefinite pentru lungime si latime, nu doar random

# Ex: a)
# class RandomRectangle:
# 	# TODO: implementarea voastra aici 

# with RandomRectangle() as my_rectangle:
# 	print(f”Area of rectangle is: <val_area>”)

# b) 
# with RandomRectangle() as my_rectangle:   # random 
# 	print(f”Area of rectangle is: <val_area>”)

# with RandomRectangle(3, 4) as my_rectangle:  # predef
# 	print(f”Area of rectangle is: <val_area>”)
# 	# => Area of rectangle is: 12


# a)
import random

# v1
print("a)\nv1: ")
class RandomRectange():
    def __enter__(self):
        self.width = random.randrange(1,10)
        self.length = random.randrange(1,10)
        print(f"Rectangle(width = {self.width}, length = {self.length})")
        return self

    def area(self):
        return self.width * self.length

    def __exit__(self, type, value, traceback):
        pass

with RandomRectange() as my_rectangle:
    print(f'The area is {my_rectangle.area()}')

    
# v2
print("\nv2: ")
class RandomRectange():
    def __enter__(self):
        self.width = random.randrange(1,10)
        self.length = random.randrange(1,10)
        print(f"Rectangle(width = {self.width}, length = {self.length})")
        return [self.width, self.length]   # returnam o lista formata din cele doua valori

    def __exit__(self, type, value, traceback):
        pass

with RandomRectange() as my_rectangle:
    print(f'The area is {my_rectangle[0] * my_rectangle[1]}')   # pe care le putem accesa prin index in interiorul context managerului
    

    
# b) valori predefinite, nu random
print("\nb)")
class RandomRectange():
    def __init__(self, width=None, length=None):
        self.width = width
        self.length = length

    def __enter__(self):
        if self.width is None:
            self.width = random.randrange(1,10)
        if self.length is None:
            self.length = random.randrange(1,10)
        print(f"Rectangle(width = {self.width}, length = {self.length})")
        return self 

    def area(self):
        return self.width * self.length

    def __exit__(self, type, value, traceback):
        pass
    
with RandomRectange(3,4) as my_rectangle: 
    print(f'The area is {my_rectangle.area()}')
```

```python
# 2. Fiind date urmatoarele functii, scrieti un decorator tva_decorator pt cele 2 functii care sa returneze pretul produsului cu TVA adaugat (tva = 19%).

# def get_tv_price(price):
#     return price

# def get_laptop_price(price):
#     return price

# Ex: print(get_tv_price(2000)) # => 2380


def tva_decorator(original_func):
    def wrapper_func(*args):
        initial_price = original_func(*args)
        final_price = 0.19 * initial_price + initial_price
        print(f"Price with TVA added is {final_price}")
        return final_price
    
    
    return wrapper_func

@tva_decorator
def get_tv_price(price):
    print(f"Price of TV is: {price}")
    return price

@tva_decorator
def get_laptop_price(price):
    print(f"Price of laptop is: {price}")
    return price


print(get_tv_price(2000))
print(get_laptop_price(4800))
```
