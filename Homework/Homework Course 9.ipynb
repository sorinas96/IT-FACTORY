```python
# 1.Given the multi-line string student_grades:
#  a). Using the below data, create your student_grades.csv from student_grades string.
#  b). List the contents of the file as a list of dicts, a row being a dict of form:
#  {"student_name": <name_of_student>, "subject": <name_of_subject>, "grade": <grade>}
#  c). List the student that has the most grades and how many it has   # => Celine, 3  
# poate fi si un dict {"student": "Celine", "no_of_grades": 3} sau {"Celine": 3} sau ("Celine", 3), "Student with most grades.."

# student_grades ="""
# Eveline,math,9
# Andrew,physics,6
# Celine,music,10
# Matthew,math,7
# Celine,english,10
# Matthew,english,10
# Celine,chemistry,6
# """

student_grades ="""
Eveline,math,9
Andrew,physics,6
Celine,music,10
Matthew,math,7
Celine,english,10
Matthew,english,10
Celine,chemistry,6
"""

# a)
import csv
rows_as_strings = student_grades.strip("\n").split("\n")  # facem split dupa \n (rand nou) si obtinem o lista de stringuri
# unde fiecare string reprezinta un rand din csv-ul final
rows = []
for row in rows_as_strings:
    rows.append(row.split(','))  # similar cu citirea cand rezultatul era o lista de liste, si aici avem nevoie de aceeasi 
    # structura, asa ca transformam fiecare rand dintr`un string intr`o lista de stringuri, unde fiecare string e o coloana a randului
    
print(f"a) rows to write: {rows}")
with open("student_grades.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)
    
    
# b)
# v1
print("\nb) v1:")
students_list = []
with open("student_grades.csv", "r", newline="") as f:
    reader = csv.reader(f)
    
    for row in reader:
        students_list.append({"student_name": row[0], "subject": row[1], "grade": row[2]})
print(students_list)

# v2     
print("\nb) v2:")
students_list = []
with open("student_grades.csv", "r", newline="") as f:
    reader = csv.DictReader(f, fieldnames=["student_name", "subject", "grade"])
    
    for row in reader:
        students_list.append(row)
print(students_list)
        
        
# c)
#  v1 
print("\nc) v1:")

no_of_grades = {}
# ne construim dictionarul no_of_grades de forma {<nume_student1>: <numar_de_note1>, ...}
for student in students_list:
    key = student['student_name']
    if key in no_of_grades:   
        no_of_grades[student['student_name']] += 1
    else:
        no_of_grades[student['student_name']] = 1

most_grades = max(no_of_grades.values())      # aflam numarul maxim de note;
for student in no_of_grades:                  # cautam cheia cu numele studentului
    if no_of_grades[student] == most_grades:  # care are numarul maxim de note
        student_no_of_grades = {   # ne construim un dictionar, ca sa arate bine rezultatul :)
            "student": student,
            "no_of_grades": most_grades
        }
print(student_no_of_grades)

#  v2
print("\nc) v2:")

from collections import defaultdict
no_of_grades = defaultdict(int)   # default dict ne permite sa definim o valore default daca o cheie
# nu exista in dict, in cazul nostru vrem ca valoarea default sa fie cea aferenta int, adica 0

# ne construim dictionarul no_of_grades de forma {<nume_student1>: <numar_de_note1>, <nume_student2>: <numar_de_note2>, ...}
for student in students_list:
    no_of_grades[student['student_name']] += 1   # asa ca putem incrementa cheia cu numele studentului
                                                # chiar daca nu o avem inca in dictionar
        
student_with_most_grades = max(no_of_grades, key=lambda x: no_of_grades[x]) # aflam numele studentului
# cu cele mai multe note (adica cheia din dict no_of_grades care are valoarea cea mai mare)
student_no_of_grades = {          # ne construim dictionarul final
    "student": student_with_most_grades, 
    "no_of_grades": no_of_grades[student_with_most_grades]
}
print(student_no_of_grades)
```

```python
import json
#  a). Write a function that will print all the details of a person. The function will receive an id, 
#  ex: print(show_details_by_id("aeec8c5e-f513-4dfd-8435-1d49eb90877c")) => 
# {
#          "first_name": "Mike",
#          "last_name": "McMan",
#          "sex": "M",
#          "age": 52,
#          "partner": "d48da772-fdb0-4b14-9ff5-aafe29c15dc9",
#          "children": ["dd69e184-ee18-491c-ba36-db625ba35d8b", "b14031ee-bec5-4902-8e78-95508bd89b64", "9fc00e67-5d00-4830-8679-4e2e66f7fe64"]
#  }
# v1
print("\na) v1:")
def show_details_by_id_v1(person_id):
    with open('people.json', 'r') as f:
        people_dict = json.load(f)
        
    return people_dict['people'][person_id]

print(show_details_by_id_v1("aeec8c5e-f513-4dfd-8435-1d49eb90877c"))


# v2 - codul structurat ca o aplicatie scalabila  
print("\na) v2:")
FILE_PATH = "people.json"  # presupunem ca fisierul se afla in directorul curent, daca nu, va trebui modificat PATH-ul corespunzator

def read_database(file_path=FILE_PATH):
    with open(file_path, "r") as f:
        people = json.load(f)
    return people["people"]


def show_details_by_id_v2(person_id):
    people_dict = read_database()
    return people_dict[person_id]

print(show_details_by_id_v2("aeec8c5e-f513-4dfd-8435-1d49eb90877c"))


# v3 - optimizam utilizarea memoriei adaugan un argument optional functiei show_details_by_id
print("\na) v3:")
def show_details_by_id_v3(person_id, people_dict=None):
    if people_dict is None:
        people_dict = read_database()
    return people_dict[person_id]

people_dict = read_database()  # citim dinainte baza de date
print(show_details_by_id_v3("aeec8c5e-f513-4dfd-8435-1d49eb90877c", people_dict))  # si pasam datele din ea ca argument
```

```python
# b). Write a function that will print all the details of a person in a readable way (first name, last name, age, sex, partner, children). 
# The function will receive the first_name and last_name of the person.
# e.g.: show_details_by_first_and_last_name("Mike", "McMan") => First name: Mike
#                                                               Last name McMan
#                                                               Age: 52
#                                                               Sex: M
#                                                               Partner: Mary McMan
#                                                               Children: ['Anna', 'Adrian', 'Danna']
# v1
print("\nb) v1:")
def show_details_by_first_and_last_name_v1(first_name, last_name):
    people_dict = read_database() # singurul loc in care citim baza de date
    for person_id, person_details in people_dict.items():
        if person_details['first_name'] == first_name and person_details['last_name'] == last_name:
            partner = people_dict[person_details['partner']]  # putem sa returnam direct detaliile partenerului sau... (v2)
            children = [people_dict[child_id]['first_name'] for child_id in person_details['children']]
            print(f"First name: {person_details['first_name']}\n"
                  f"Last name: {person_details['last_name']}\n"
                  f"Age: {person_details['age']}\n"
                  f"Sex: {person_details['sex']}\n"
                  f"Partner: {partner['first_name']} {partner['last_name']}\n"
                  f"Children: {children}"
                  )

show_details_by_first_and_last_name_v1("Mike", "McMan")


# v2 
print("\nb) v2:")
def show_details_by_first_and_last_name_v2(first_name, last_name, people_dict=None):
    if people_dict is None:
        people_dict = read_database()  

    for person_id, person_details in people_dict.items():
        if person_details['first_name'] == first_name and person_details['last_name'] == last_name:
            partner = show_details_by_id_v3(person_details['partner'], people_dict)  # evident ca acesta e un exemplu trivial si nu e nevoie neaparata sa apelam functia show_details_by_id_v3 
            children = [show_details_by_id_v3(child_id, people_dict)['first_name'] for child_id in person_details['children']]
            print(f"First name: {person_details['first_name']}\n"
                  f"Last name: {person_details['last_name']}\n"
                  f"Age: {person_details['age']}\n"
                  f"Sex: {person_details['sex']}\n"
                  f"Partner: {partner['first_name']} {partner['last_name']}\n"
                  f"Children: {children}"
                  )

people_dict = read_database()
show_details_by_first_and_last_name_v2("Mike", "McMan")
```

```python
# e). List the details of the children 'Mike' and 'Mary' have together.

def show_details_of_common_children(id_person1, id_person2, people_dict=None):
    if people_dict is None:
        people_dict = read_database() 
    
    parent1_children_list = set(people_dict[id_person1]["children"])
    parent2_children_list = set(people_dict[id_person2]["children"])
    common_children_ids = parent1_children_list.intersection(parent2_children_list)
    for child_id in common_children_ids:
        print(show_details_by_id_v3(child_id, people_dict))


show_details_of_common_children("aeec8c5e-f513-4dfd-8435-1d49eb90877c", "d48da772-fdb0-4b14-9ff5-aafe29c15dc9", people_dict)
```

```python
#  f). Add a function to marry 2 people, for example Anna to John and do the required changes! 
def read_database_v2(file_path=FILE_PATH):
    with open(file_path, "r") as f:
        people = json.load(f)
#     return people["people"]
    return people

def write_database(data_dict, file_path=FILE_PATH):
    with open(file_path, "w") as f:
        json.dump(data_dict, f, indent=4)


def marry(id_person1, id_person2, db=None):
    if db is None:
        db = read_database_v2()
    people_dict = db['people']
    
    if id_person1 in people_dict and id_person2 in people_dict:  # ne asiguram ca ambele id-uri se afla in DB
        if people_dict[id_person1]['sex'] == 'F':   # schimbam numele de familie in cazul fetei
            people_dict[id_person1]['last_name'] = people_dict[id_person2]['last_name']
        else:
            people_dict[id_person2]['last_name'] = people_dict[id_person1]['last_name']

        people_dict[id_person1]['partner'] = id_person2  # le adaugam unul altuia id-ul partenerului
        people_dict[id_person2]['partner'] = id_person1

        db['people'] = people_dict  # puteti sa omiteti linia asta, fiindca people_dict pointeaza catre db, db se va modifica odata cu people_dict
        write_database(db)   
        print(f"\n\nDB FILE AFTER: {json.dumps(db, indent=4)}")
    else:
        print("We could not find the 2 persons in DB.")
        

marry("dd69e184-ee18-491c-ba36-db625ba35d8b", "711f79a9-8ec8-45d4-8de6-76ae06c7b7cd")  # marry Anna & John
```
